```python
#Libraries
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#Dataset
df = pd.DataFrame({'from':['John', 'Anne', 'Sara', 'Bob', 'Caren', 'Danny'], 'to':['Anne', 'Sara', 'Bob', 'Danny', 'John' ,'Caren' ]})
G=nx.from_pandas_edgelist(df, 'from', 'to')
```

```python
#Graph
nx.draw(G, with_labels=True, node_color='pink', edge_color='blue')
plt.show()
```
