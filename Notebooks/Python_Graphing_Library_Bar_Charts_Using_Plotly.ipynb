```python
#Libraries
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
```

```python
#Dataset
data_UnitedKingdom = px.data.gapminder().query("country == 'United Kingdom'")
```

```python
#Graph
fig = px.bar(data_UnitedKingdom, x='year', y='pop')
fig.show()
```

```python
#Graph
fig = px.bar(data_UnitedKingdom, x='year', y='pop',
             hover_data=['lifeExp', 'gdpPercap'], color='lifeExp',
             labels={'pop':'population of: United Kingdom'}, height=400)
fig.show()
```

```python
#Dataset
days = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat','Sun']
```

```python
#Graph
fig = go.Figure()
fig.add_trace(go.Bar(
    x=days,
    y=[20, 14, 25, 16, 18, 22, 19],
    name='Primary Product',
    marker_color='lightblue'))

fig.add_trace(go.Bar(
    x=days,
    y=[10, 24, 35, 26, 38, 12, 29],
    name='Secondary Product',
    marker_color='lightsalmon'))

fig.show()
```

```python
#Graph
colors = ['lightslategray',] * 7
colors[1] = 'lightsalmon'

fig = go.Figure(data=[go.Bar(
    x=['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat','Sun'],
    y=[20, 14, 25, 16, 18, 22, 19],
    marker_color=colors # marker color can be a single color value or an iterable
)])
fig.update_layout(title_text='Least Profitable day')


fig.show()
```

```python
#Dataset
labels = ["Python","JavaScript","C# ","Java"]
widths = np.array([10,20,20,50])

data = {
    "South": [50,80,60,70],
    "North": [50,20,40,30],
    "East": [50,80,60,70],
    "West": [50,20,40,30],

}
```

```python
#Graph
fig = go.Figure()
for key in data:
    fig.add_trace(go.Bar(
        name=key,
        y=data[key],
        x=np.cumsum(widths)-widths,
        width=widths,
        offset=0,
        customdata=np.transpose([labels, widths*data[key]]),
        texttemplate="%{y} x %{width} =<br>%{customdata[1]}",
        textposition="inside",
        textangle=0,
        textfont_color="white",
        hovertemplate="<br>".join([
            "label: %{customdata[0]}",
            "width: %{width}",
            "height: %{y}",
            "area: %{customdata[1]}",
        ])
    ))

fig.update_xaxes(
    tickvals=np.cumsum(widths)-widths/2,
    ticktext= ["%s<br>%d" % (l, w) for l, w in zip(labels, widths)]
)

fig.update_xaxes(range=[0,200])
fig.update_yaxes(range=[0,200])

fig.update_layout(
    title_text="Marimekko Chart",
    barmode="stack",
    uniformtext=dict(mode="hide", minsize=10),
)
```

```python
#Dataset
years = ['2016','2017','2018', '2019', '2020', '2021', '2022']
```

```python
#Graph
fig = go.Figure()
fig.add_trace(go.Bar(x=years, y=[300, 400, 500, 600, 700, 800, 900],
                base=[-300, -400, -500, -600, -700, -800, -900],
                marker_color='lightsalmon',
                name='expenses'))
fig.add_trace(go.Bar(x=years, y=[100, 200, 300, 400, 500, 700, 800],
                base=0,
                marker_color='lightslategrey',
                name='revenue'
                ))

fig.show()
```
