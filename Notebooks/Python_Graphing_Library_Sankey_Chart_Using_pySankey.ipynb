```python
#Libraries
import pandas as pd
from pySankey.sankey import sankey
```

```python
#Dataset
url = "https://raw.githubusercontent.com/anazalea/pySankey/master/pysankey/fruits.txt"
df = pd.read_csv(url, sep=" ", names=["true", "predicted"])

colors = {
    "apple": "#ffe6e6",
    "blueberry": "#ff9999",
    "banana": "#ff4d4d",
    "lime": "#cc0000",
    "orange": "#660000"
}
```

```python
#Graph
sankey(df["true"], df["predicted"], aspect=20, colorDict=colors, fontsize=15)
```
