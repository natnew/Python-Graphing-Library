```python
#Libraries
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
```

```python
#Dataset
df = px.data.gapminder().query("country=='United States'")
fig = px.line(df, x="year", y="lifeExp", title='Life expectancy in: United States')
```

```python
#Graph
fig.show()
```

```python
#Export as a static image (Formats: .png/.jpeg/.webp/.svg/.pdf/.eps)
fig.write_image("images/LinePlot_Canada.png")
```

```python
#Dataset
df = px.data.gapminder().query("continent=='Americas'")
fig = px.line(df, x="year", y="lifeExp", title='Life expectancy in: Americas ', color='country')
```

```python
#Graph
fig.show()
```

```python
#Export as a static image (Formats: .png/.jpeg/.webp/.svg/.pdf/.eps)
fig.write_image("images/LinePlot_Americas.png")
```

```python
#Dataset
df = px.data.gapminder().query("country in ['Canada', 'United Kingdom']")
```

```python
#Graph
fig = px.line(df, x="lifeExp", y="gdpPercap", color="country", text="year")
fig.update_traces(textposition="bottom right")
fig.show()
```

```python
#Export as a static image (Formats: .png/.jpeg/.webp/.svg/.pdf/.eps)
fig.write_image("images/ConnectedLinePlot_CanadaUnitedKingdom.png")
```

```python
#Graph
fig = px.line(df, x="lifeExp", y="gdpPercap", color="country", text="year", symbol="country")
fig.update_traces(textposition="bottom right")
fig.show()
```

```python
#Export as a static image (Formats: .png/.jpeg/.webp/.svg/.pdf/.eps)
fig.write_image("images/ConnectedLinePlot_CanadaUnitedKingdom_symbols.png")
```

```python
#Dataset
df = px.data.stocks()
```

```python
#Graph
fig = px.line(df, x='date', y=["AAPL", "MSFT"])
fig.show()
```

```python
#Export as a static image (Formats: .png/.jpeg/.webp/.svg/.pdf/.eps)
fig.write_image("images/LinePlot_Stock.png")
```

```python
#Graph
fig.update_layout(
    showlegend=False,
    plot_bgcolor="white",
    margin=dict(t=10,l=10,b=10,r=10)
)

fig.show()
```
