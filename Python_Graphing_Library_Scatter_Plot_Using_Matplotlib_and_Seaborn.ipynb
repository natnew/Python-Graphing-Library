```python
#Libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Dataset
#Data can be found here: https://github.com/mwaskom/seaborn-data
df = sns.load_dataset('planets')
```

```python
#Points
sns.regplot(x=df["mass"], y=df["distance"], color='pink')
plt.show()
```
